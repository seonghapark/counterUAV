```python
binary_file = open('test_binary.txt','r') # 연속된 sync 개수를 새는 코드
```

```python
total_list = binary_file.readlines()
```

```python
len(total_list)
```

```python
isZero = True
count = 0
total_count = 0
for i in range(1000):
    count += 1
    #print(total_list[i], end='')
    if '0b000' in total_list[i] and isZero == False:
        print('연속된 1 sync: ', count) 
        total_count += count
        count = 0
        isZero = True
        continue
    elif '0b001' in total_list[i] and isZero == True:
        print('연속된 0 sync: ', count) 
        total_count += count
        count = 0
        isZero = False
        continue
    
```
